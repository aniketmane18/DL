```python
sequence_length = 5
data = np.array([i for i in range(100)])
x_data = []
y_data = []
for i in range(len(data) - sequence_length):
    x_data.append(data[i:i+sequence_length])
    y_data.append(data[i+sequence_length])

x_data = np.array(x_data)
y_data = np.array(y_data)
```

```python
x_data = x_data.reshape((x_data.shape[0], x_data.shape[1], 1))
```

```python
train_size = int(len(x_data) * 0.8)
x_train, x_test = x_data[:train_size], x_data[train_size:]
y_train, y_test = y_data[:train_size], y_data[train_size:]
```

```python
model = Sequential([
    SimpleRNN(10, activation='tanh', input_shape=(sequence_length, 1)),
    Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x_train, y_train, epochs=20, batch_size=8, validation_data=(x_test, y_test))
```

```python
loss = model.evaluate(x_test, y_test)
print(f"Test Loss: {loss}")
```

```python
# Calculate accuracy
def calculate_accuracy(predictions, actuals, tolerance=0.1):
    # Tolerance is the acceptable percentage deviation from the actual value
    accurate_predictions = np.abs(predictions - actuals) <= tolerance * np.abs(actuals)
    accuracy = np.mean(accurate_predictions) * 100  # Convert to percentage
    return accuracy

# Calculate and print accuracy
accuracy = calculate_accuracy(predictions.flatten(), y_test, tolerance=0.1)
print(f"Model Accuracy: {accuracy:.2f}%")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.metrics import MeanAbsolutePercentageError

# Generate sample data
# Let's create a simple sequence dataset: predicting the next value in a sequence
# For example, given [0, 1, 2, 3], the target is [1, 2, 3, 4]
sequence_length = 5
data = np.array([i for i in range(100)])

# Normalize the data to the range [0, 1]
data = (data - np.min(data)) / (np.max(data) - np.min(data))

x_data = []
y_data = []
for i in range(len(data) - sequence_length):
    x_data.append(data[i:i+sequence_length])
    y_data.append(data[i+sequence_length])

x_data = np.array(x_data)
y_data = np.array(y_data)

# Reshape data to fit RNN input format (samples, time steps, features)
x_data = x_data.reshape((x_data.shape[0], x_data.shape[1], 1))

# Split into training and testing sets
train_size = int(len(x_data) * 0.8)
x_train, x_test = x_data[:train_size], x_data[train_size:]
y_train, y_test = y_data[:train_size], y_data[train_size:]

# Build the RNN model
model = Sequential([
    SimpleRNN(50, activation='tanh', input_shape=(sequence_length, 1)),
    Dense(1)
])

# Compile the model
model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.001), 
              loss='mse', 
              metrics=[MeanAbsolutePercentageError()])

# Train the model
model.fit(x_train, y_train, epochs=100, batch_size=8, validation_data=(x_test, y_test))

# Evaluate the model
loss, mape = model.evaluate(x_test, y_test)
print(f"Test Loss: {loss}")
print(f"Mean Absolute Percentage Error (MAPE): {mape}")

# Make predictions
predictions = model.predict(x_test)
print("Predictions:", predictions[:5].flatten())
print("Actual:", y_test[:5])

# Calculate accuracy
def calculate_accuracy(predictions, actuals, tolerance=0.1):
    # Tolerance is the acceptable percentage deviation from the actual value
    accurate_predictions = np.abs(predictions - actuals) <= tolerance * np.abs(actuals)
    accuracy = np.mean(accurate_predictions) * 100  # Convert to percentage
    return accuracy

# Calculate and print accuracy
accuracy = calculate_accuracy(predictions.flatten(), y_test, tolerance=0.1)
print(f"Model Accuracy: {accuracy:.2f}%")

# Example usage
example_input = np.array([95, 96, 97, 98, 99])  # A sequence to predict the next value
example_input = (example_input - np.min(data)) / (np.max(data) - np.min(data))  # Normalize example input
example_input = example_input.reshape((1, sequence_length, 1))
example_prediction = model.predict(example_input)
print("Example Input:", example_input.flatten())
print("Predicted Next Value:", example_prediction.flatten()[0])
```
